```python
from google.colab import files
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
```

```python
import PyPDF2

def read_resume(file_path):
    text = ""
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text()
    return text

print(read_resume(file_name))
```

```python
from google.colab import userdata
```

```python
from google import genai
client = genai.Client(api_key=userdata.get('GOOGLE_API_KEY'))
```

```python
def analyze_resume(text):
    response = client.models.generate_content(
        model="gemini-2.5-flash-lite",
        contents= f"""
You are an ATS resume analyzer.
Analyze the resume below and return ATS score and feedback.
Resume:
{text}
Return in JSON format with: - ats_score (0-100) - summary - strengths - weaknesses
Rules: - Output ONLY JSON - No extra text
"""
    )
    return response.text
```

```python
resume_text = read_resume(file_name)
result = analyze_resume(resume_text)
print(result)
```

```python
import json

data = json.loads(result)

print("ATS Score :", data['ats_score'])
print("Summary :", data['summary'])

print("\nStrengths")
for s in data['strengths']:
    print("-", s)

print("\nWeaknesses")
for w in data['weaknesses']:
    print("-", w)
```
